```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import random


plt.rcParams['figure.figsize'] = [12, 12]
```

```python
fp_red = r'Data/AVIATE 15July21 50ft half image size_mosaic_red.tif'
fp_nir = r'Data/AVIATE 15July21 50ft half image size_mosaic_nir.tif'

RED = np.transpose(rasterio.open(fp_red).read(),(1,2,0))
NIR = np.transpose(rasterio.open(fp_nir).read(),(1,2,0))

RED_VECT = np.reshape(RED, [-1,1])
NIR_VECT = np.reshape(NIR, [-1,1])
```

```python
tmp = RED_VECT.copy()
tmp = tmp[RED_VECT != 0]

NIR_VECT = NIR_VECT[RED_VECT != 0]
RED_VECT = tmp

tmp = NIR_VECT.copy()
tmp = tmp[NIR_VECT != 0]

RED_VECT = RED_VECT[NIR_VECT != 0]
NIR_VECT = tmp
```

```python
idx=random.sample(range(len(RED_VECT)),100000)
plt.scatter(RED_VECT[idx]/(2**16),NIR_VECT[idx]/(2**16))
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].hist(RED_VECT)
ax[1].hist(NIR_VECT)
plt.show()
```

```python
idx=random.sample(range(len(RED_VECT)),100000)
# Generate fake data
x = RED_VECT[idx]
y = NIR_VECT[idx]

# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)


plt.scatter(x, y, c=z, s=100)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```
